```python
# Импортируем библиотеки
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
df = pd.read_csv("C:/study/intelect/train.csv")
```

```python
df.fillna({'Age': df['Age'].median(),  'Embarked': df['Embarked'].mode()[0]}, inplace=True)
# для возраста использовал медиану, из-за устойчивости к выбросу, embarked - стобец из категории, логичнее использовать самое частое 
df.drop('Cabin', axis=1, inplace=True)
df.drop('Ticket', axis=1, inplace=True)
df.drop('Name', axis=1, inplace=True)
# дропнут Cabin ибо имела много пропущенных значений - являлась неинформативной
```

```python
df.isnull().sum()
```

```python


scaler = MinMaxScaler()
columns_for_scale = ['Age', 'Fare']
df[columns_for_scale] = scaler.fit_transform(df[columns_for_scale])
```

```python
df.head()
```

```python
columns_for_encode = ['Sex', 'Embarked']
df = pd.get_dummies(df, columns=columns_for_encode, drop_first=True)
```
