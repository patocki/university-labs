```python
import os
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib
matplotlib.use('Agg')  # Используем бэкенд без GUI
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Подготовка данных для обучения модели
# Генерация тестовых данных
np.random.seed(42)
X = np.random.randint(0, 2, size=(100, 12))  # 100 примеров, 12 бинарных признаков
Y = np.random.randint(0, 2, size=(100, 2))  # 100 примеров, 2 класса (one-hot encoding)

# Сохранение данных в файлы
np.savetxt('dataIn.txt', X, fmt='%d')
np.savetxt('dataOut.txt', Y, fmt='%d')

# Загрузка данных
X = np.loadtxt('dataIn.txt')
Y = np.loadtxt('dataOut.txt')

# Разделение на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

print(f"Размер обучающей выборки: {X_train.shape}")
print(f"Размер тестовой выборки: {X_test.shape}")

# 2. Создание нейросети с одним скрытым слоем и функцией активации logsig
model = keras.Sequential([
    keras.layers.Input(shape=(12,)),  # Явное указание входного слоя
    keras.layers.Dense(10, activation='sigmoid'),  # Скрытый слой с logsig
    keras.layers.Dense(2, activation='softmax')    # Выходной слой
])

# Компиляция модели
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 3. Обучение модели на предоставленных данных
print("Начало обучения...")
history = model.fit(X_train, y_train,
                    epochs=100,
                    batch_size=8,
                    validation_data=(X_test, y_test),
                    verbose=1)

# 4. Проверка точности на тестовых примерах
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Точность на тестовой выборке: {test_accuracy:.4f}")

# Предсказания на тестовой выборке
y_pred = model.predict(X_test)
y_pred_class = np.argmax(y_pred, axis=1)
y_true_class = np.argmax(y_test, axis=1)

print("\nОтчет классификации:")
print(classification_report(y_true_class, y_pred_class,
                          target_names=['Правящая партия', 'Оппозиция']))

# 5. Визуализация результатов предсказаний (сохранение в файлы)
# График изменения функции ошибки
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('График функции ошибки')
plt.legend()
plt.grid(True)
plt.savefig('loss_plot.png', dpi=300, bbox_inches='tight')
plt.close()

# График изменения точности
plt.figure(figsize=(10, 4))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('График точности')
plt.legend()
plt.grid(True)
plt.savefig('accuracy_plot.png', dpi=300, bbox_inches='tight')
plt.close()

# Матрица ошибок
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_true_class, y_pred_class)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Правящая партия', 'Оппозиция'],
            yticklabels=['Правящая партия', 'Оппозиция'])
plt.xlabel('Предсказанный класс')
plt.ylabel('Истинный класс')
plt.title('Матрица ошибок')
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.close()

print("\nМатрица ошибок:")
print(cm)
print(f"\nОбщая точность: {test_accuracy:.2%}")
print("\nГрафики сохранены в файлы:")
print("- loss_plot.png")
print("- accuracy_plot.png")
print("- confusion_matrix.png")

# Дополнительная информация
print(f"\nКоличество примеров в тестовой выборке: {len(y_test)}")
print(f"Примеров класса 'Правящая партия': {np.sum(y_true_class == 0)}")
print(f"Примеров класса 'Оппозиция': {np.sum(y_true_class == 1)}")
```
